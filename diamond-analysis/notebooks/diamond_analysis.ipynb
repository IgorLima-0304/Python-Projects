```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

file_path = r'C:\Users\Estudo-Trabalho\Desktop\Python-Projects\diamond-analysis\data\diamonds.csv'
df = pd.read_csv(file_path)

print(df.head())

df['cut'] = df['cut'].astype('category').cat.codes
df['color'] = df['color'].astype('category').cat.codes
df['clarity'] = df['clarity'].astype('category').cat.codes

plt.figure(figsize=(12, 6))
sns.histplot(df['price'] / 1000, bins=50, kde=True, color='blue')
plt.xlabel('Pre√ßo (mil d√≥lares)')
plt.ylabel('Frequ√™ncia')
plt.title('Distribui√ß√£o dos Pre√ßos dos Diamantes')
plt.show()

print("""
üìå **Gr√°fico 1 - Distribui√ß√£o dos Pre√ßos dos Diamantes**
Este histograma mostra como os pre√ßos dos diamantes est√£o distribu√≠dos. A maioria dos diamantes custa menos de **5 mil d√≥lares**.
Podemos ver um **vi√©s positivo**, ou seja, h√° alguns diamantes extremamente caros que aumentam a m√©dia do pre√ßo.
""")

plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['carat'], y=df['price'] / 1000, alpha=0.5, color='red')
plt.xlabel('Peso (Carat)')
plt.ylabel('Pre√ßo (mil d√≥lares)')
plt.title('Peso (Carat) vs Pre√ßo')
plt.show()

print("""
üìå **Gr√°fico 2 - Peso (Carat) vs Pre√ßo**
Este gr√°fico mostra a rela√ß√£o entre o **peso do diamante (em carats)** e o **pre√ßo**.
Existe uma rela√ß√£o clara: **diamantes maiores s√£o mais caros**. No entanto, h√° muita variabilidade de pre√ßos,
o que sugere que outros fatores (como qualidade do corte e cor) tamb√©m influenciam no valor.
""")

plt.figure(figsize=(12, 6))
sns.boxplot(x=df['cut'], y=df['price'] / 1000, palette='coolwarm')
plt.xlabel('Qualidade do Corte (Codificado)')
plt.ylabel('Pre√ßo (mil d√≥lares)')
plt.title('Distribui√ß√£o dos Pre√ßos por Tipo de Corte')
plt.show()

print("""
üìå **Gr√°fico 3 - Distribui√ß√£o dos Pre√ßos por Tipo de Corte**
O corte do diamante afeta seu pre√ßo. Este boxplot mostra a **mediana** (linha dentro da caixa), 
os **quartis** e os **outliers** (pontos fora das caixas).  
Os cortes de melhor qualidade tendem a ser **mais valorizados**, mas h√° varia√ß√£o significativa em todas as categorias.
""")

plt.figure(figsize=(10, 5))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mapa de Calor das Correla√ß√µes')
plt.show()

print("""
üìå **Gr√°fico 4 - Mapa de Calor das Correla√ß√µes**
Este mapa de calor mostra **quais vari√°veis t√™m rela√ß√£o entre si**.  
O **pre√ßo** tem uma forte correla√ß√£o com o **peso (carat)**, o que confirma o gr√°fico anterior.  
Al√©m disso, algumas vari√°veis como **x, y e z** (dimens√µes do diamante) tamb√©m est√£o fortemente correlacionadas.
""")

plt.figure(figsize=(12, 6))
sns.countplot(x=df['color'], palette='viridis')
plt.xlabel('Categoria de Cor (Codificada)')
plt.ylabel('Quantidade')
plt.title('Distribui√ß√£o de Diamantes por Categoria de Cor')
plt.show()

print("""
üìå **Gr√°fico 5 - Distribui√ß√£o de Diamantes por Categoria de Cor**
Este gr√°fico mostra quantos diamantes existem em cada categoria de cor.  
A cor pode afetar o pre√ßo, pois diamantes mais brancos e sem colora√ß√£o s√£o mais valiosos.  
A categoria **G** parece ser a mais comum no dataset.
""")

print("‚úÖ **An√°lise conclu√≠da com sucesso!**")
```
