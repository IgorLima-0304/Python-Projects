```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

file_path = r'C:\Users\Estudo-Trabalho\Desktop\Python-Projects\diamond-analysis\data\diamonds.csv'
df = pd.read_csv(file_path)

print(df.head())

df['cut'] = df['cut'].astype('category').cat.codes
df['color'] = df['color'].astype('category').cat.codes
df['clarity'] = df['clarity'].astype('category').cat.codes

plt.figure(figsize=(12, 6))
sns.histplot(df['price'] / 1000, bins=50, kde=True, color='blue')
plt.xlabel('Preço (mil dólares)')
plt.ylabel('Frequência')
plt.title('Distribuição dos Preços dos Diamantes')
plt.show()

print("""
📌 **Gráfico 1 - Distribuição dos Preços dos Diamantes**
Este histograma mostra como os preços dos diamantes estão distribuídos. A maioria dos diamantes custa menos de **5 mil dólares**.
Podemos ver um **viés positivo**, ou seja, há alguns diamantes extremamente caros que aumentam a média do preço.
""")

plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['carat'], y=df['price'] / 1000, alpha=0.5, color='red')
plt.xlabel('Peso (Carat)')
plt.ylabel('Preço (mil dólares)')
plt.title('Peso (Carat) vs Preço')
plt.show()

print("""
📌 **Gráfico 2 - Peso (Carat) vs Preço**
Este gráfico mostra a relação entre o **peso do diamante (em carats)** e o **preço**.
Existe uma relação clara: **diamantes maiores são mais caros**. No entanto, há muita variabilidade de preços,
o que sugere que outros fatores (como qualidade do corte e cor) também influenciam no valor.
""")

plt.figure(figsize=(12, 6))
sns.boxplot(x=df['cut'], y=df['price'] / 1000, palette='coolwarm')
plt.xlabel('Qualidade do Corte (Codificado)')
plt.ylabel('Preço (mil dólares)')
plt.title('Distribuição dos Preços por Tipo de Corte')
plt.show()

print("""
📌 **Gráfico 3 - Distribuição dos Preços por Tipo de Corte**
O corte do diamante afeta seu preço. Este boxplot mostra a **mediana** (linha dentro da caixa), 
os **quartis** e os **outliers** (pontos fora das caixas).  
Os cortes de melhor qualidade tendem a ser **mais valorizados**, mas há variação significativa em todas as categorias.
""")

plt.figure(figsize=(10, 5))
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mapa de Calor das Correlações')
plt.show()

print("""
📌 **Gráfico 4 - Mapa de Calor das Correlações**
Este mapa de calor mostra **quais variáveis têm relação entre si**.  
O **preço** tem uma forte correlação com o **peso (carat)**, o que confirma o gráfico anterior.  
Além disso, algumas variáveis como **x, y e z** (dimensões do diamante) também estão fortemente correlacionadas.
""")

plt.figure(figsize=(12, 6))
sns.countplot(x=df['color'], palette='viridis')
plt.xlabel('Categoria de Cor (Codificada)')
plt.ylabel('Quantidade')
plt.title('Distribuição de Diamantes por Categoria de Cor')
plt.show()

print("""
📌 **Gráfico 5 - Distribuição de Diamantes por Categoria de Cor**
Este gráfico mostra quantos diamantes existem em cada categoria de cor.  
A cor pode afetar o preço, pois diamantes mais brancos e sem coloração são mais valiosos.  
A categoria **G** parece ser a mais comum no dataset.
""")

print("✅ **Análise concluída com sucesso!**")
```
